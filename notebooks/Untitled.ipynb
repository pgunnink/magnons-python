```python
from scipy import linalg
import numpy as np
from magnons.energies import hamiltonian_AB
from magnons.amplitudes import AkBkAngle, AkBk
from magnons.yig import a, S, mu, J
from magnons.cgs import E_to_GHz
from magnons.magnetization import magnetization_angle
import matplotlib.pyplot as plt
```

```python
N = 100
A, B = AkBk(10**4, 10**4, N=N, J=J, S=S, h=900 * mu, eps=a**(-2), a=a, mu=mu)
mat = np.block([[A,B],[-B.T.conj(), -A]])
```

```python
E1 = np.sort(linalg.eigvals(mat) * E_to_GHz)
E2 = np.sort(linalg.eigvalsh(mat) * E_to_GHz)
```

```python
plt.plot(np.abs(E2-E1))
```
